- String slicing same as number shown before

```python
name="Omar Faruk"
print("Print name using indexing:", name[0], name[1], name[2])  # Print first three characters using indexing
print('Print last character using negative indexing:', name[-1])  # Print last character using negative indexing
print("Reversed String using slicing:", name[::-1])   # Reverse the string using slicing
print("Substring from index 2 to 6:", name[2:7])  # Extract substring from index 2 to 6
print("Substring from end to index 4:", name[:5])   # Extract substring from end to index 4
print("Substring from index 5 to end:", name[5:])   # Extract substring from index 5 to end 
print("Every second character starting from index 1:", name[1::2])  # Extract every second character starting from index 1   
print("Uppercase:", name.upper())  # Convert to uppercase
print("Lowercase:", name.lower())  # Convert to lowercase
print("Replace 'Omar' with 'Ahmed':", name.replace("Omar", "Ahmed"))  # Replace substring   
print("Split into words:", name.split(" "))  # Split the string into a list of words 
print("Length of the string:", len(name))  # Get the length of the string
print("Count of 'a':", name.count("a"))  # Count occurrences of 'a' in the string
print("Find 'Faruk':", name.find("Faruk"))  # Find the starting index of substring "Faruk"
print("Faruk" in name)  # Check if substring "Faruk" is in the string
print("John" not in name)  # Check if substring "John" is not in the string
print(name.strip())  # Remove leading and trailing whitespace
print(f"My name is {name} and I am learning Python.")  # Formatted string
print("Last element of name:", name[-1])  # Print
```
